```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

x_series = list(np.arange(0, 187))

currentDir = os.getcwd()

df = pd.read_csv("ptbdb_abnormal.csv", header=None)
df2 = pd.read_csv("ptbdb_normal.csv", header=None)
```

```python
df_key = df.pop(df.columns.values[-1])
df2_key = df2.pop(df2.columns.values[-1])
```

```python
df.tail()
```

```python
df2.tail()
```

```python
df_key
```

```python
df2_key
```

```python
for x in range (0, len(df)-1):
    row = df.iloc[x]
    num = str(int(df_key[x]))
    plt.plot(x_series, row, color='black')
    plt.grid(False)
    plt.savefig(currentDir + '\ptbdb\\' + num + '\img' + str(x) + '.png', dpi='figure', metadata=None, bbox_inches='tight', pad_inches=0, facecolor='white', edgecolor='white', backend=None)
    plt.clf()

```

```python
for x in range (0, len(df2)-1):
    row = df2.iloc[x]
    num = str(int(df2_key[x]))
    plt.plot(x_series, row, color='black')
    plt.grid(False)
    plt.savefig(currentDir + '\ptbdb\\' + num + '\img' + str(x) + '.png', dpi='figure', metadata=None, bbox_inches='tight', pad_inches=0, facecolor='white', edgecolor='white', backend=None)
    plt.clf()
```

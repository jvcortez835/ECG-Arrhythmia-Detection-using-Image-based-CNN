```python
from keras.models import load_model
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense

#from keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.utils import img_to_array, load_img

import numpy as np
```

```python
# -----------------------------------------------------------------------------
# Dimensions of images
# -----------------------------------------------------------------------------
img_width, img_height = 357, 234
input_shape = (img_width, img_height, 3)

# -----------------------------------------------------------------------------
# Load model
# -----------------------------------------------------------------------------
test_model = load_model('model.h5')

# -----------------------------------------------------------------------------
# Image data path
# -----------------------------------------------------------------------------
basedir = "testBSLRescale/"

# -----------------------------------------------------------------------------
# Prediction
# -----------------------------------------------------------------------------
i=0

def predict(basedir, i, model, r_list):
    
    path = basedir+str(i)+'.PNG'
    
    img = load_img(path,False,target_size=(img_width,img_height))
        
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
        
    preds = model.predict(x)
    #probs = model.predict_proba(x)
        
    if preds==0: print('normal')
    if preds==1: print('abnormal')
    r_list.append(preds[0][0])
    print ('')
    print ('Probability: ')
    
    return preds
```

```python
 
# ***********************************************    
# MAIN
results_0 = []
for i in range (1,51): # Images to be tested

    print('Test Sample: ', i)

    (preds)=predict(basedir, i, test_model, results_0) # prediction 

    print(' ')
```
